# Функции

```python
'''Визуализация ROC AUC и confusion matrix'''
def visualise_model(Y_test, probs, y_pred):
    
    #### ROC AUC
    fpr, tpr, threshold = roc_curve(Y_test, probs)
    roc_auc = roc_auc_score(Y_test, probs)

    fig, ax = plt.subplots(figsize=(10,5))
    plt.plot([0, 1], label='Baseline', linestyle='--')
    plt.plot(fpr, tpr, label = 'Regression')
    ax.set_title('ROC AUC = %0.3f'%roc_auc)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc='lower right')
    
    #### confusion matrix
    fig, ax = plt.subplots(figsize=(10,5))
    sns.set_context(context='paper', font_scale=2, rc=None)
    group_names = ['True Pos', 'False Pos', 'False Neg', 'True Neg']
    group_counts = ['{0:0.0f}'.format(value) for value in
                    confusion_matrix(Y_test, y_pred).flatten()]
    labels = [f'{v1}\n{v2}' for v1, v2 in
              zip(group_names, group_counts)]
    labels = np.asarray(labels).reshape(2, 2)
    ax = sns.heatmap(confusion_matrix(Y_test, y_pred),
                     annot=labels, fmt='', cmap='Greens')
    ax.set(xlabel='predicted', ylabel='real',title = 'Confusion matrix')
    plt.show()
```

# Подготовка данных к ML

```python
# обработка категориальных переменных
data = pd.get_dummies(data, columns = cat_cols)
```

```python
# Логарифмирование "хвостатых переменных" среди численных
num_cols_tails = ['income', 'decline_app_cnt', 'bki_request_cnt']

for col in num_cols_tails:
    data[col] = np.log(data[col] + 1)
```

```python
# Стандартизация численных переменных
data[num_cols] = pd.DataFrame(StandardScaler().fit_transform(data[num_cols]),
                              columns=num_cols)
```

```python
# Разделим выборку обратно на тест и трейн
train = data.query('train == 1').drop(['train'], axis=1)
test = data.query('train == 0').drop(['train'], axis=1)
```

# Наивные модели

## Как есть 
sklearn.linear_model.LogisticRegression

```python
X = train.drop(columns = ['default', 'client_id','app_date'])
Y = train['default']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                               random_state=RANDOM_SEED,
                                               test_size=0.2)

model = LogisticRegression()
model.fit(X_train, Y_train)
probs = model.predict_proba(X_test)
y_pred = model.predict(X_test)
probs = probs[:,1]

visualise_model(Y_test, probs, y_pred) # вывод ROC AUC и confusion matrix
```

Модель показала хороший ROC AUC, но она "дает кредиты всем подрят" и вызванно это сильным перекосом значений колонки default 

## Undersampling
sklearn.linear_model.LogisticRegression

```python
# Выровним количество дефолтов с количеством их отсутствия.
# Для этого случайно урежем выборку клиентов без дефолта.
under_train = pd.concat([train[train['default'] == 0].sample(
    len(train[train['default'] == 1]), random_state=RANDOM_SEED),
                         train[train['default'] == 1]])

X = under_train.drop(columns=['default', 'client_id', 'app_date'])
Y = under_train['default']

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, random_state=RANDOM_SEED, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
probs = model.predict_proba(X_test)
probs = probs[:, 1]

visualise_model(Y_test, probs, y_pred)
```

ROC AUC немного снизился, но зато модель теперь с $\frac{2}{3}$ вероятностью определяет как дефолт так и его отсутсвие, что уже неплохо

## Гиперпараметры, Undersampling
sklearn.linear_model.LogisticRegression

```python
# Попробуем поменять class_weight и с помощью GridSearchCV найти оптимальные параметры
lr_best = LogisticRegression(class_weight='balanced',
                            random_state=RANDOM_SEED)

parameters_grid = {'C': np.linspace(0, 30, 10),
         'max_iter': np.linspace(10,300,10, dtype=int)}

model_best = GridSearchCV(lr_best, parameters_grid, verbose=1, cv=3)

model_best.fit(X_train, Y_train)
probs = model_best.predict_proba(X_test)
y_pred = model_best.predict(X_test)
probs = probs[:,1]

visualise_model(Y_test, probs, y_pred)
```

Модель с другими параметрами стала немного хуже, подбор параметров показал, что изначальная настройка достаточно оптимальна.

## Undersampling
sklearn.linear_model._stochastic_gradient.SGDClassifier

```python
model = SGDClassifier(max_iter=1000, tol=1e-9,loss='log', random_state=RANDOM_SEED)

model.fit(X_train, Y_train)
probs = model.predict_proba(X_test)
y_pred = model.predict(X_test)
probs = probs[:,1]


visualise_model(Y_test, probs, y_pred)
```

Эта модель отказывает больше клиентам, которые могут выплатить кредит, нежели тем, кто потерпел дефолт

## Undersampling
statsmodels.api.Logit

```python
model = Logit(Y_train,X_train).fit()

probs = model.predict(X_test)
y_pred = list(map(round, model.predict(X_test)))

visualise_model(Y_test, probs, y_pred)
```

Эта модель себя показала лучше всего, оставлю её для каггла.
